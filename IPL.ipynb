```python
# Extracting the records where a team won by batting first
batting_first = ipl[ipl['WonBy'] == 'Runs']
```

```python
# Extracting the records where a team won by batting first
batting_first = ipl[ipl['WonBy'] == 'Runs']
```

```python
# Finding the number of wins w.r.t each team after batting first
display(batting_first['WinningTeam'].value_counts())
```

```python
# Making bar plot of top 3  with most wins after batting first
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['WinningTeam'].value_counts()[0:3].keys()),list(batting_first['WinningTeam'].value_counts()[0:3]),color=["Red","blue","green"])
plt.show()
```

```python
# Making a pie-chart distribution
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['WinningTeam'].value_counts()),labels=list(batting_first['WinningTeam'].value_counts().keys()),autopct = '%0.1f%%')
plt.show()
```
